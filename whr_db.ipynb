```python
# Import modules
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Pull data from pgAmin4 to import data 
engine = create_engine('postgresql://postgres:postgres@localhost:5432/WHR_DB')
conn = engine.connect()

#Or pull in from CSV file
whr_sql = "Resources/whr_db.csv"
```

```python
# Read from sql file
# whr = pd.read_sql("SELECT * FROM whr_db", conn)

# # Read from CSV file
whr = pd.read_csv(whr_sql)

whr
```

```python
# Convert into dataframe and sort from best to worst overall ladder score
whr_db = pd.DataFrame(whr)
whr_db = whr_db.sort_values('ladder_score', ascending=False)
whr_db
```

```python
# Who had the highest and lowest WHR score over time?
```

```python
# Group by country and sort from best to worst overall ladder score
groupby_country = whr_db.groupby(['country_name']).mean()

whr_countrydb = groupby_country.sort_values('ladder_score', ascending=False)

whr_countrydb = whr_countrydb.drop(columns = ["year"])

whr_countrydb
```

```python
# Who had the highest and lowest WHR score in 2021?
```

```python
# Review of the top 10 in 2021
whr_2021 = whr_db.loc[whr_db['year']== 2021].head(10)
whr_2021
```

```python
# Review of the bottom 10 in 2021
whr_2021 = whr_db.loc[whr_db['year']== 2021].tail(10)
whr_2021
```

```python
#COVID Analysis
```

```python
# Who had the most COVID deaths per population?
whr_covid = whr_countrydb.dropna(subset = ["covid_deaths"])
whr_covid["covid_deaths_per_total_deaths"]=whr_covid["covid_deaths"]/whr_covid["total_deaths"]
whr_covid["covid_deaths_per_population"] = whr_covid["covid_deaths"]/whr_covid["population"]

whr_covid = whr_covid.drop(columns=['ladder_score','log_gdp_per_capita','social_support','healthy_life_expectancy','freedom_to_make_life_choices','generosity','perceptions_of_corruption','total_deaths','population'])

whr_covid = whr_covid.sort_values('covid_deaths_per_total_deaths', ascending=False)

whr_covid
```

```python
# Creating columns for 2017, 2018, 2019, 2020, 2021

whr_2021= whr_db.loc[whr_db['year']== 2021]
whr_2021 = whr_2021[["country_name","ladder_score"]]
whr_2021.columns = ['country_name', '2021']
covidvwhr1 = pd.merge(whr_covid, whr_2021, on = 'country_name')

whr_2020= whr_db.loc[whr_db['year']== 2020]
whr_2020 = whr_2020[["country_name","ladder_score"]]
whr_2020.columns = ['country_name', '2020']
covidvwhr2 = pd.merge(covidvwhr1, whr_2020, on = 'country_name')

whr_2019= whr_db.loc[whr_db['year']== 2019]
whr_2019 = whr_2019[["country_name","ladder_score"]]
whr_2019.columns = ['country_name', '2019']
covidvwhr3 = pd.merge(covidvwhr2, whr_2019, on = 'country_name')

whr_2018 = whr_db.loc[whr_db['year']== 2018]
whr_2018 = whr_2018[["country_name","ladder_score"]]
whr_2018.columns = ['country_name', '2018']
covidvwhr4 = pd.merge(covidvwhr3, whr_2018, on = 'country_name')

whr_2017= whr_db.loc[whr_db['year']== 2017]
whr_2017 = whr_2017[["country_name","ladder_score"]]
whr_2017.columns = ['country_name', '2017']
covidvwhr5 = pd.merge(covidvwhr4, whr_2017, on = 'country_name')

whr_2016= whr_db.loc[whr_db['year']== 2016]
whr_2016 = whr_2016[["country_name","ladder_score"]]
whr_2016.columns = ['country_name', '2016']
covidvwhr6 = pd.merge(covidvwhr5, whr_2016, on = 'country_name')

covidvwhr = covidvwhr6[["country_name","2016","2017","2018","2019","2020","2021","covid_deaths_per_total_deaths","covid_deaths_per_population"]]
covidvwhr
```
